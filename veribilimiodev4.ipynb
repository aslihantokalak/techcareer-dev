```python

#Techcareer Ödev 4

#1. soru

#Numpy ile Matris İşlemleri
#5x5 boyutunda rastgele (0-100 arasında) tam sayılardan oluşan bir matris oluşturun.
#Bu matrisin:
 #Ortalama, standart sapma ve varyansını hesaplayın.
#En büyük ve en küçük değerlerini bulun.
#Köşegenindeki elemanların toplamını bulun.

import numpy as np

# 5x5 boyutunda 0-100 arasında rastgele tam sayı matrisi
matris = np.random.randint(0, 101, (5, 5))
print("Matris:\n", matris)

# Ortalamayı bulurken yine np kütüphanesi
ortalama = np.mean(matris)
print("Ortalama:", ortalama)

# Standart Sapma
std_sapma = np.std(matris)
print("Standart Sapma:", std_sapma)

# Varyans
varyans = np.var(matris)
print("Varyans:", varyans)

# En büyük değeri bul
en_buyuk = np.max(matris)
print("En Büyük Değer:", en_buyuk)

# En küçük değeri bul
en_kucuk = np.min(matris)
print("En Küçük Değer:", en_kucuk)

# Köşegenlerin toplamı
kosegen_toplam = np.trace(matris)
print("Köşegen Elemanlarının Toplamı:", kosegen_toplam)

#2. soru

#2. Numpy ile Veri Simülasyonu
#1000 adet öğrencinin sınav puanlarını (0-100 arası, normal dağılımdan gelen) simüle edin.
#Ortalama, medyan ve standart sapmayı hesaplayın.
#50’den düşük alan kaç öğrenci olduğunu bulun.


import numpy as np

# 1000 öğrencinin sınav puanlarını normal dağılım simülasyonu
# Ortalama = 50, Standart Sapma = 15
puanlar = np.random.normal(loc=50, scale=15, size=1000)

# Puanları 0-100 arası
puanlar = np.clip(puanlar, 0, 100)

# Ortalaması
ortalama = np.mean(puanlar)
print("Ortalama:", ortalama)

# Medyanı
medyan = np.median(puanlar)
print("Medyan:", medyan)

# Standart Sapması
std_sapma = np.std(puanlar)
print("Standart Sapma:", std_sapma)

# 50’den düşük alan öğrenci sayısı
elli_altinda = np.sum(puanlar < 50)
print("50’den düşük alan öğrenci sayısı:", elli_altinda)

#3. soru
#Bir CSV dosyası oluşturun ya da aşağıdaki gibi örnek bir DataFrame hazırlayın:

#	Görevler:
#	1. Veri setini Pandas DataFrame olarak oluşturun.
#	2. Her ders için ortalama puanı bulun.
#	3. En yüksek matematik notunu alan öğrenciyi bulun.
#	4. Her öğrencinin not ortalamasını hesaplayan yeni bir sütun ekleyin.
#	5. Bölümlere göre gruplayarak her bölümün ortalama başarılarını hesaplayın.
#	6. Ortalaması 70’in üzerinde olan öğrencileri filtreleyin.



data ={
    "Öğrenci": ["Ali", "Ayşe", "Mehmet", "Zeynep", "Ahmet"],
"Yaş": [20, 21, 19, 22, 20],
"Bölüm": ["Bilgisayar", "Fizik", "Kimya", "Bilgisayar", "Fizik"],
"Matematik": [70, 60, 80, 90, 55],
"Fizik": [65, 75, 70, 85, 60],
"Kimya": [80, 85, 65, 95, 70]
}
import pandas as pd
df = pd.DataFrame(data)
print(df)   

# 3.2 Her ders için ortalama puanlar
print("Her dersin ayrıca ortalaması:\n", df[["Matematik", "Fizik", "Kimya"]].mean(), "\n")

# 3.3 En yüksek Matematik notunu alan öğrenci
en_yuksek_matematik = df.loc[df["Matematik"].idxmax(), "Öğrenci"]
print("En yüksek Matematik notunu alan öğrenci:", en_yuksek_matematik, "\n")
import numpy as np

# 3.4 Her öğrencinin not ortalamasını hesaplayan  sütun
df["Ortalama"] = df[["Matematik", "Fizik", "Kimya"]].mean(axis=1)
print("Not ortalamaları eklenmiş DataFrame:\n", df, "\n")

#3.5 Bölümlere göre ortalama başarı sütunu
print("Bölümlere göre ortalama başarı:\n", df.groupby("Bölüm")[["Matematik", "Fizik", "Kimya", "Ortalama"]].mean(), "\n")

# 3.6 Ortalaması 70’in üzerinde olan öğrencilerin listesi
print("Ortalaması 70’in üzerinde olan öğrenciler:\n", df[df["Ortalama"] > 70])

#4. soru

# Matplotlib veya Seaborn kullanımı
# Her dersin dağılımını histogram ile gösterin.
# Bölümlere göre ortalamaları bar grafikte görsel olarak ifade edilmesi

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veri setini tekrar oluşturalım
data = {
    "Öğrenci": ["Ali", "Ayşe", "Mehmet", "Zeynep", "Ahmet"],
    "Yaş": [20, 21, 19, 22, 20],
    "Bölüm": ["Bilgisayar", "Fizik", "Kimya", "Bilgisayar", "Fizik"],
    "Matematik": [70, 60, 80, 90, 55],
    "Fizik": [65, 75, 70, 85, 60],
    "Kimya": [80, 85, 65, 95, 70]
}

df = pd.DataFrame(data)

# Her bir öğrencinin ortalaması
df["Ortalama"] = df[["Matematik", "Fizik", "Kimya"]].mean(axis=1)

# - Her dersin dağılımını histogram ile gösterilmesi
plt.figure(figsize=(12,4))

for i, ders in enumerate(["Matematik", "Fizik", "Kimya"], 1):
    plt.subplot(1, 3, i)
    sns.histplot(df[ders], bins=5, kde=True, color="skyblue")
    plt.title(f"{ders} Dağılımı")

plt.tight_layout()
plt.show()

# - Bölümlere göre ortalamaları bar grafikte gösterilmesi
plt.figure(figsize=(6,4))
sns.barplot(x="Bölüm", y="Ortalama", data=df, ci=None, palette="viridis")
plt.title("Bölümlere Göre Ortalama Başarı")
plt.ylabel("Ortalama Puan")
plt.show()
```
