```python
#1. Soru - Birden Fazla İşlem İçeren Fonksiyonlara Giriş
# Tekrarları silip benzersiz hale getiren fonksiyon
def benzersiz_notlar(notlar):
    return list(set(notlar))

# En yüksek notu bulan fonksiyon
def en_yuksek_not(notlar):
    return max(notlar)

# En düşük notu bulan fonksiyon
def en_dusuk_not(notlar):
    return min(notlar)

# Notları küçükten büyüğe sıralayan fonksiyon
def sirala_notlar(notlar):
    return sorted(notlar)


# Ana kod bloğu
notlar = [85, 92, 76, 92, 100, 76, 85, 92]

# Önce benzersiz hale getiriyoruz
benzersiz = benzersiz_notlar(notlar)

print("Benzersiz Notlar:", benzersiz)
print("En Yüksek Not:", en_yuksek_not(benzersiz))
print("En Düşük Not:", en_dusuk_not(benzersiz))
print("Sıralı Notlar:", sirala_notlar(benzersiz))

#2. Soru Armstrong Sayısı Kontrolü

def armstrong_check(sayi):
    # Sayıyı stringe çevirerek basamaklara ayırıyoruz
    basamaklar = str(sayi)
    basamak_sayisi = len(basamaklar)
    
    # Her basamağın basamak_sayisi kadar kuvvetini alıp topluyoruz
    toplam = sum(int(b)**basamak_sayisi for b in basamaklar)
    
    # Armstrong sayısı kontrolü
    return toplam == sayi


# Kullanıcıdan sayı al
sayi = int(input("Bir sayı girin: "))

# Kontrol et ve sonucu yazdır
if armstrong_mu(sayi):
    print(sayi, "bir Armstrong sayısıdır.")
else:
    print(sayi, "bir Armstrong sayısı değildir.")

#3  Küme İşlemleri
# Kümeler
# Ortak dilleri bulan fonksiyon
def ortak_diller(A, B):
    return A & B

# Sadece A'da olan dilleri bulan fonksiyon
def sadece_A_diller(A, B):
    return A - B

# Birleşimi bulan ve alfabetik sıralayan fonksiyon
def birlesim_alfabetik(A, B):
    return sorted(A | B)


# Ana program
A = {"Python", "R", "SQL", "Java"}
B = {"C++", "Python", "JavaScript", "SQL"}

print("Ortak Diller:", ortak_diller(A, B))
print("Sadece A’da Olan Diller:", sadece_A_diller(A, B))
print("Birleşim (alfabetik):", birlesim_alfabetik(A, B))

#4 Modüller

# Rastgele sayılar ve istatistik modüllerini içe aktar

import random
import statistics

def rastgele_sayilar(n, baslangic=1, bitis=100):
    # n tane rastgele sayı üret
    sayilar = [random.randint(baslangic, bitis) for _ in range(n)]
    
    # Ortalama ve standart sapma
    ortalama = statistics.mean(sayilar)
    std_sapma = statistics.stdev(sayilar)
    
    return sayilar, ortalama, std_sapma


# Fonksiyonu çağırma (1–100 arasında 10 sayı)
sayilar, ort, std = rastgele_sayilar(10, 1, 100)

print("Rastgele Sayılar:", sayilar)
print("Ortalama:", ort)
print("Standart Sapma:", std)

#5 Kelime Sayacı

def kelime_sayaci(metin):
    # Noktalama işaretlerini kaldırmak için
    for noktalama in [".", ",", "!", "?", ";", ":"]:
        metin = metin.replace(noktalama, "")
    
    # Metni kelimelere ayır
    kelimeler = metin.split()
    
    # 1. Toplam kelime sayısı
    toplam = len(kelimeler)
    
    # 2. En uzun kelime
    en_uzun = max(kelimeler, key=len)
    
    # 3. En sık geçen kelime
    frekans = {}
    for k in kelimeler:
        frekans[k] = frekans.get(k, 0) + 1
    en_sik = max(frekans, key=frekans.get)
    
    return toplam, en_uzun, en_sik


# Test edelim
metin = "Python yaygın bir dil üstelik kolayca anlaşılır. Python öğrenmek çok eğlenceli!"
toplam, en_uzun, en_sik = kelime_sayaci(metin)

print("Toplam kelime sayısı:", toplam)
print("En uzun kelime:", en_uzun)
print("En sık geçen kelime:", en_sik)

sayilar = [5, 12, 7, 18, 24, 3, 16]

# 6 Filtreleme, Dönüştürme ve Sıralama

def islem_yap(liste):
    # Sadece çift sayıları filtreleme
    cift_sayilar = filter(lambda x: x % 2 == 0, liste)
    
    # Çift sayıların karelerini bulma
    kareleri = map(lambda x: x**2, cift_sayilar)
    
    # Kareleri azalan sırada sıralama
    sirali_liste = sorted(kareleri, reverse=True)
    
    return sirali_liste

sonuc = islem_yap(sayilar)
print(f"Çift sayıların kareleri azalan sırada: {sonuc}")


# 7 Lambda Fonksiyonları ve Sıralama

kelimeler = ["veri", "bilim", "analiz", "yapayzeka", "python"]

def kelimeleri_sirala(liste):
  # sorted fonksiyonu ile kelimeleri uzunluklarına göre sıralama
  # key=lambda kelime: len(kelime) ifadesi, sıralama kriterini belirler
  sirali_liste = sorted(liste, key=lambda kelime: len(kelime))
  return sirali_liste

sirali_kelimeler = kelimeleri_sirala(kelimeler)
print(f"Kelimeler uzunluklarına göre sıralanmış hali: {sirali_kelimeler}")

# 8 Metodlar

def rakamlari_topla(metin):
    toplam = 0
    gecici_sayi = ""
    
    for karakter in metin:
        # Karakter bir rakam mı diye kontrol et
        if karakter.isdigit():
            gecici_sayi += karakter
        else:
            # Rakam olmayan bir karaktere denk gelince,
            # birikmiş olan sayıyı toplama ekle
            if gecici_sayi:
                toplam += int(gecici_sayi)
                gecici_sayi = ""
    
    # Döngü bittikten sonra stringin sonunda bir sayı kalmış olabilir.
    # Bu son sayıyı da toplama ekle.
    if gecici_sayi:
        toplam += int(gecici_sayi)
        
    return toplam

# Ödevdeki kullanım:
metin1 = "abc12def3"
sonuc1 = rakamlari_topla(metin1)
print(f'"{metin1}" stringinin rakamlarının toplamı: {sonuc1}') # Çıktı: 15

# 9 Numpy Kütüphanesi

import numpy as np

# 0-50 arasında 10 rastgele tam sayıdan oluşan bir dizi oluşturma
dizi = np.random.randint(0, 51, 10)

print(f"Oluşturulan dizi: {dizi}")

# Dizinin ortalamasını bulma
ortalama = np.mean(dizi)

# Dizinin standart sapmasını bulma
standart_sapma = np.std(dizi)

# Dizinin en büyük değerini bulma
en_buyuk_deger = np.max(dizi)

print("-" * 30)
print(f"Dizinin ortalaması: {ortalama:.2f}")
print(f"Dizinin standart sapması: {standart_sapma:.2f}")
print(f"Dizinin en büyük değeri: {en_buyuk_deger}")


# 10 Numpy ile Matris İşlemleri

import numpy as np

# 5x5 boyutunda 0-1 arasında rastgele değerlerden oluşan bir matris oluşturma
np.random.seed(42)  # Tekrarlanabilir sonuçlar için seed belirleme
matris = np.random.rand(5, 5)

print("Orijinal Matris:")
print(matris)
print("-" * 30)

# Her sütunun ortalamasını bulma
sutun_ortalamalari = np.mean(matris, axis=0)

print("Her Sütunun Ortalaması:")
print(sutun_ortalamalari)
print("-" * 30)

# Matrisi binary matrise dönüştürme (0.5'ten büyükse 1, değilse 0)
binary_matris = (matris > 0.5).astype(int)

print("Binary Matris (0.5'ten büyükler 1, diğerleri 0):")
print(binary_matris)

```
